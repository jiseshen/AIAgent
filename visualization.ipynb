```python
import matplotlib.pyplot as plt
```

```python
update_interval = [10, 20, 40]
test_accuracy = [0.176, 0.0, 0.036]
ood_accuracy = [0.194, 0.0, 0.016]
test_rouge = [0.190, 0.074, 0.148]
ood_rouge = [0.220, 0.102, 0.180]

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
l1 = ax1.plot(update_interval, test_accuracy, 'g-', label="Test Acc")
l2 = ax1.plot(update_interval, ood_accuracy, 'g--', label="OOD Acc")
l3 = ax2.plot(update_interval, test_rouge, 'b-', label="Test ROUGE")
l4 = ax2.plot(update_interval, ood_rouge, 'b--', label="OOD ROUGE")
ax1.set_xlabel('Update Interval')
ax1.set_ylabel('Accuracy')
ax2.set_ylabel('ROUGE')
plt.legend(l1+l2+l3+l4, [l.get_label() for l in l1+l2+l3+l4])
plt.show()
```
